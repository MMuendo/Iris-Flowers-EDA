```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
#load the dataset
iris  = pd.read_csv(r"C:\Users\User\seaborn-data\iris.csv")
```

```python
#Display the first few rows
iris.head()
```

```python
#Summary statistics
print(iris.describe())
```

```python
# check the number of samples in each class
iris.species.value_counts()
```

```python
# Pairwise scatter plot colored by species
sns.pairplot(iris, 
             hue = 'species',
             markers = '*')
```

```python
# boxplot of the feature
plt.figure (figsize = (8,6))
sns.boxplot(iris)
plt.title('Iris Flowers Boxplot')
plt.xlabel('Iris Flowers Attributes')
plt.show()
```

```python
iris_corr = iris[['sepal_length', 'sepal_width', 'petal_length', 'petal_width']].corr()
```

```python
#heatmap
plt.figure(figsize = (8,6))
sns.heatmap(iris_corr,
            annot = True,
            cmap = 'RdBu',
            vmin = -1,
            vmax = 1)
plt.title('Iris Flowers Attributes Heatmap')
plt.show()
```

```python
#histogram
iris.hist(edgecolor='black', figsize=(10, 6), color='magenta')
plt.show()
```

```python
#violin
plt.figure(figsize = (12,6))
sns.violinplot(data = iris, 
               x='species', 
               y='sepal_length',
               palette = 'Set1')
plt.title('Distribution of Sepal length by species')
plt.xlabel('Species')
plt.ylabel('Sepal length')
plt.show()
```

```python
#swarmplot
plt.figure(figsize = (12,6))
sns.swarmplot(data = iris, 
               x='species', 
               y='petal_length',
               palette = 'Set1')
plt.title('Distribution of Sepal length by species')
plt.xlabel('Species')
plt.ylabel('Sepal length')
plt.show()
```

```python
#KDEplot
plt.figure(figsize = (12,6))
sns.kdeplot(data=iris, 
            x='sepal_width', 
            y='petal_length', 
            hue='species', 
            fill=True)
plt.title('Pairwise Density Plot: Sepal Width vs. Petal Length')
plt.xlabel('Sepal Width')
plt.ylabel('Petal Length')
plt.show()
```
